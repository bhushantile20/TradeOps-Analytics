# Exploratory Data Analysis - BTCUSD Forecasting

This notebook provides a basic outline for analyzing the BTCUSD dataset fetched from Binance.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yaml

# Load data
with open('../config.yaml', 'r') as f:
    config = yaml.safe_load(f)

df = pd.read_parquet('../' + config['data']['processed_path'])
df.head()
```

## 1. Price Over Time

```python
plt.figure(figsize=(12, 6))
plt.plot(df['timestamp'], df['close'])
plt.title('BTCUSD Price (1h)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

## 2. Target Variable Analysis (Next Return)

```python
sns.displot(df['next_return'], kde=True)
plt.title('Distribution of Next Log Returns')
plt.show()
```

## 3. Correlation Matrix

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.show()
```
